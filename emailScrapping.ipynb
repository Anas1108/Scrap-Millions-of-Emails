```python
def extract_email_from_data(text):
    match = re.findall(r'[\w\.-]+@[\w\.-]+',text)
    return match
```

```python
#extracting_all_comments_URL
def extracting_all_comments_URL(url):
    try:
        all_pages_links=[]
        driver.get(url)
        a_tag=driver.find_elements_by_css_selector('a.topictitle')
        for i in a_tag:
            all_pages_links.append(i.get_attribute('href'))
        return all_pages_links
    except:
        print("Error in extracting all comments url")
        return
# extracting_all_comments_URL("https://www.fredmiranda.com/forum/board/10")
```

```python
def further_comment_sub_pages_link(url):
    all_pages_links=[]
    driver.get(url)
    try:
        tbody_tag=(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr[2]/td[2]/table[1]/tbody/tr/td/table/tbody/tr/td/table[3]/tbody/tr/td[1]/table/tbody"))
        all_a_tag = tbody_tag.find_elements_by_tag_name('a')
        totalPages=int(all_a_tag[len(all_a_tag)-2].text)
        print((totalPages))
        pageNmbr=1
        for i in range(totalPages-1):
            pageUrl=url + '/'+str(pageNmbr+i)
            all_pages_links.append(pageUrl)
        return(all_pages_links)
    except:
        return None
# further_comment_sub_pages_link("https://www.fredmiranda.com/forum/topic/330885")
```

```python
def scrap_year_links():
    links_to_scrap_data=[]
    print("Extracting all pages url from year page...................")
    current_year_all_pages_link=all_comment__pages_of_year()
    if current_year_all_pages_link is not None:
        for i in range(len(current_year_all_pages_link)):
            print("Extracting all comments from page " + str(i+1)+"................")
            all_comments_url=extracting_all_comments_URL(current_year_all_pages_link[i])
            links_to_scrap_data.append(all_comments_url)
    return(links_to_scrap_data)
```
